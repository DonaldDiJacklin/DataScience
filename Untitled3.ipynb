```python
import numpy as np
from MyML import *
```

```python
X = np.random.randn(1000,2)
r = np.sqrt(X[:,0]**2 + X[:,1]**2)
x1 = X[r>1.4,]
x2 = X[r<1,]
y1 = np.ones(len(x1)).reshape(len(x1),1)
y2 = np.zeros(len(x2)).reshape(len(x2),1)
X = np.row_stack((x1,x2))
Y = np.row_stack((y1,y2))
```

```python
model = NeuralNetwork(indims = 2,nodes = [6,4,1], activations = ['tan','tan'], taskType = 'c')
```

```python
model.train(X,Y, epochs = 10,gamma = .8, moment = .8,lr = .005, optim = 'adam')
```

```python
model = NeuralNetwork(indims = 2,nodes = [4,6,1], activations = ['prelu','prelu'], taskType = 'c')
model.train(X,Y, epochs = 10,gamma = .8, moment = .8,lr = .005, optim = 'rms')
```

```python
model = NeuralNetwork(indims = 2,nodes = [6,10,10,1], activations = ['relu','tan','relu'], taskType = 'c')
model.train(X,Y, epochs = 10,gamma = .8, moment = .8,lr = .005, optim = 'ada')
```
