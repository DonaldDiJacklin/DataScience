```python
import numpy as np
from MyML import *
```

```python
x1 = np.random.randn(500,2)
x2 = np.random.randn(500,2)*2+10
```

```python
x = np.row_stack((x1,x2))
x = (x-np.min(x))/(np.max(x)-np.min(x))
y1 = np.ones((500,1))
y2 = np.zeros((500,1))
y = np.row_stack((y1,y2))
X = np.column_stack((x,y))
```

```python
model = TSNNrms(indims = 2, nodes = [1], taskType='c')
```

```python
model.train(X[:,0:2],np.matrix(X[:,2]).T, epochs = 100,lr =  .19, gamma = .9, mu = .9)
```

```python
accuracy(model.prediction,np.matrix(X[:,2]).T)
```
