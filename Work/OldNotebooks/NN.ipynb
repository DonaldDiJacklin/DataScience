```python
import numpy as np
import matplotlib.pyplot as plt
from Phtekitlear import NeuralNet
```

```python
X=np.linspace(0,10,(1000))
Xa=np.linspace(0,30,(1000))
X=np.column_stack((X,Xa))
a = np.matrix([[0.5,1.2,2.5],[1,3.2,1.7]])
b = np.matrix([[1,2.2,0.8],[0.9,5.5,0.7]])
c = np.matrix([4,2,7])
d= np.random.randn(1000,1)*np.matrix([100,200,150])
y = X@a+X**2@b+c+d
y=np.eye(y.shape[1])[np.argmax(np.array(y),axis=1)]
sum(y[2])
```

```python
a = [4,3,3]
model = NeuralNet.NeuralNet(indims=2,listnodes = a)
model.weightInitialization()
model.weights
```

```python
model.predict(X)
```

```python
model.weightInitialization()
model.train(X, y, epochs=100,lr=.001)
```

```python
model.predict(X)
```

```python
ypred=np.eye(model.predict(X).shape[1])[np.argmax(model.predict(X),axis=1)]
```
