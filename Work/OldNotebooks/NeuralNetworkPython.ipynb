```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh(X):
    return (np.exp(X)-np.exp(-X))/(np.exp(X)+np.exp(-X))
```

```python
r = np.linspace(-100,100,1000)
y = tanh(r)
```

```python
plt.plot(r,y)
```

```python
X = np.random.randn(500,2)
```

```python
phi = np.column_stack((np.ones(500),X))
```

```python
w1 = np.random.randn(3,6)/10
A = phi@w1
```

```python
z = tanh(A)
zprime = np.column_stack((np.ones(500),z))
```

```python
w2 = np.random.randn(7,3)/10000
A2 = zprime@w2
```

```python
z2 = tanh(A2)
z2prime = np.column_stack((np.ones(500),z2))
```

```python
w3 = np.random.randn(4,1)/10000
A3 = z2prime@w3
```

```python
def sig(X):
    return 1/(1+np.exp(-X))
```

```python
Y = np.round(sig(A3))
```

```python
plt.hist(Y)
```
