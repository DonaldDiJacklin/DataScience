```r
install.packages("scatterplot3d")
library(scatterplot3d)

X <- cbind(rnorm(3000),(1:3000)/3000)
Phi <- cbind(rep(1,3000),X)

b <- c(1,2,3)

Y <- Phi%*%b + rnorm(3000)


w <- c(2,3,4)

```

```r
scatterplot3d(Y,X[,1],X[,2])
```

```r
error<-c()
r<- solve(t(Phi)%*%Phi)%*%t(Phi)%*%Y
dim(r)
dim(Phi)
dim(t(Phi))
dim(Y)
for(i in 1:2000){
    print(t(Phi%*%w - Y)%*%(Phi%*%w - Y))
    w <- w - .0001*t(Phi)%*%(Phi%*%w - Y)
}
```

```r
1-t(Y - Phi%*%w)%*%(Y - Phi%*%w)/(t(Y - mean(Y))%*%(Y - mean(Y)))
```

```r
newPhi = cbind(rep(1,3000),rnorm(3000))
newb = c(.1,.2)
newY = newPhi%*%newb + .01*rnorm(3000)

neww = c(1,1)
```

```r
for(i in 1:200){
#     print(t(Phi%*%neww - Y)%*%(Phi%*%neww - Y))
    neww <- neww - .0001*t(newPhi)%*%(newPhi%*%neww - newY)
}
print(neww)
1-t(Y - newPhi%*%neww)%*%(Y - newPhi%*%neww)/(t(Y - mean(Y))%*%(Y - mean(Y)))
```

```r
F = rnorm(3000*100)
dim(F) = c(3000,100)
newerPhi = cbind(newPhi, F)
```
