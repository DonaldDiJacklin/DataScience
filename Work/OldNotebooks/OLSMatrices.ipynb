```python
# Imports the requisite packages.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constructs two Independent variables.
X1 = np.array(np.linspace(0,2,200)).T +np.array(
    np.random.uniform(2,25,200)).T
X2 = np.array(np.random.uniform(5,30,200)).T

# Turns the inputs into [1 , X_11 , X_21]
#                       [2 , X_12 , X_22]
#                        ...
X = np.column_stack((np.ones(200).T, X1, X2))

# Constructs the true Beta
b = np.array([[2],[3],[10]])

# Spits out Y values.
Y = np.matmul(X,b) + np.array(np.random.normal(0,1,(200,1)))*3

# Makes a 3D plot of X1, X2, and Y.
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(X1,X2,Y)
```

```python
# Solves the system of equations XTXw = XTY
w = np.linalg.solve(np.dot(X.T,X),np.dot(X.T,Y))
print(w)
```

```python
# The following computes R-squared
Xw = np.dot(X,w)
Top = np.dot((Y - Xw).T, (Y - Xw))
diffrommean = Y - np.mean(Y)
Bottom = np.dot((diffrommean.T),diffrommean)
R2 = 1 - Top/Bottom
R2
```
