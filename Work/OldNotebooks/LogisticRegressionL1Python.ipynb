```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.randn(1000,100)
```

```python
X = np.column_stack((np.ones(1000),X))
```

```python
H = 1  + 2*X[:,1] - 3*X[:,2]
```

```python
Y = np.rint(1/(1+np.exp(-H)))
```

```python

x1 = np.linspace(-3,3,1000)
x2 = 1/3 + 2*x1/3

colors = np.where(Y == 1, 'y', 'k')
plt.scatter(X[:,1],X[:,2], c = colors)
plt.scatter(x1,x2)
```

```python
def sig(matrix):
    return 1/(1+np.exp(-matrix))
X = np.array(X)
Y = np.matrix(Y).T
w = np.matrix(np.random.randn(101)).T
for i in range(0,3000):
    w = w - .000001*X.T@(sig(X@w) - Y)
```

```python
nx2 = -np.asscalar(w[0]/w[2]) - np.asscalar(w[1]/w[2])*x1
```

```python
# colors = np.where(np.array(Y).tolist() == 1, 'y', 'k')
X1 = X[:,1].tolist()
X2 = X[:,2].tolist()
plt.scatter(X1,X2, c = colors)
plt.scatter(x1,nx2)
```

```python
def sig(matrix):
    return 1/(1+np.exp(-matrix))

w = np.matrix(np.random.randn(101)).T
for i in range(0,3000):
    w = w - .5*X.T@(sig(X@w) - Y) - .0000001*np.matrix(np.sign(w))
```

```python
nx22 = -np.asscalar(w[0]/w[2]) - np.asscalar(w[1]/w[2])*x1
X1 = X[:,1].tolist()
X2 = X[:,2].tolist()
plt.scatter(X1,X2, c = colors)
plt.scatter(x1,nx22)
```
