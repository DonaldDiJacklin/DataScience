```python
import numpy as np
import matplotlib.pyplot as plt
X = np.linspace(0,10,10)
X

Y = 3 + 4*X + 3*np.random.randn(10)
Y[8:10] = Y[8:10] + 50

new = np.column_stack((np.ones(10),X))

w = np.linalg.inv(new.T.dot(new)).dot(new.T).dot(Y)

print(w)

plt.scatter(X,Y)
plt.plot(X, w[0] +w[1]*X)
```

```python
wReg = np.linalg.inv(new.T.dot(new) + 200*np.identity(2)).dot(new.T).dot(Y)
plt.scatter(X,Y)
plt.plot(X, wReg[0] + wReg[1]*X)
print(wReg)
```
