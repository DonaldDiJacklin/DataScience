```python
import numpy as np
from MyML import *
```

```python
X = np.random.randn(1500,2)*5
r = np.sqrt(X[:,0]**2 + X[:,1]**2)
```

```python
Ys = np.zeros(1500)
for i in range(0,1500):
    if r[i]>5:
        Ys[i] = 1
    if r[i]<=5:
        Ys[i] = 0
```

```python
s = Ys == 0
t = Ys == 1
```

```python
Y = np.matrix(np.column_stack((s,t)).astype(int))
Y = Y[:,0]
```

```python
model = TSNN(indims = 2,nodes = [4,4,3,1], taskType = 'c')
```

```python
model.train(X,Y, epochs = 3)
```

```python
newmodel = TSNN(indims = 2, nodes = [500,500,1], activation = 'relu',taskType = 'c')
```

```python
newmodel.train(X,Y, epochs = 3)
```
