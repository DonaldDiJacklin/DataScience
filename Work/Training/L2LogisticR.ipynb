```r
x1<-rnorm(3000,0,1)
x1[10:30]<-x1[10:30]+100
x2 <- rnorm(3000,0,2)
x2[10:30]<-x2[10:30]+100
X<- cbind(x1,x2, rep(1,3000))

y1 <- rnorm(3000, 1,.2)
y2 <- rnorm(3000, 1, .9)
y2[10:30]<-y2[10:30] - 100

y<- cbind(y1,y2,rep(0,3000))

```

```r
plot(X[,1],X[,2], col = 'red',ylim = c(-100,100))
points(y[,1],y[,2], col = 'black')
```

```r
phi <- rbind(X,y)[,1:2]
for(i in 1:2){
    phi[,i]<-(phi[,i] - min(phi[,i]))/(max(phi[,i]) - min(phi[,i]))
}
Y <- rbind(X,y)[,3]
```

```r
phi<- cbind(rep(1,3000),phi)
```

```r
sig<- function(matrix){
    return(1/(1+exp(-matrix)))
}
w<-as.matrix(rnorm(3))
for(i in 1:20000){
    w<-w - .00001*t(phi)%*%((sig(phi%*%w)) -Y)
}
```

```r
x1s <- -10:100
x2s<- -(w[1] +w[2]*x1s)/w[3]
```

```r
plot(phi[,2],phi[,3], col = factor(Y))
# points(y[,1],y[,2], col = 'black')
lines(x1s,x2s)
```
