```python
X1 = np.random.randn(1000,2)
X2 = np.random.randn(1000,2)*1.5+5
y1 = np.ones(1000)
y2 = np.zeros(1000)
full = np.row_stack((np.column_stack((X1,y1)),np.column_stack((X2,y2))))
np.random.shuffle(full)
plt.plot(X1[:,0],X1[:,1])
plt.plot(X2[:,0],X2[:,1])
```

```python
newstochastic = TSNN(indims = 2,nodes = [10,5,1])
newstochastic.train(full[:,0:2],full[:,2:],mu = 0, vocal = True)
zerrs = newstochastic.errs
```

```python
newstochastic = TSNN(indims = 2,nodes = [10,5,1])
newstochastic.train(full[:,0:2],full[:,2:],mu = .1)
tenerrs = newstochastic.errs
```
