```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stat
```

```python
X = abs(np.random.randn(10000))*10
```

```python
Y = X**2 + 3 + np.random.randn(10000)*50
Y[3:10] = Y[3:10]+1000
Y[36:45] = Y[36:45] - 1000
```

```python
plt.scatter(X,Y)
```

```python
newX = np.linspace(-5,50,200)
```

```python
def dist(a,b):
    return (a-b)**2
```

```python
D = np.empty((len(newX),len(X)))
for i in range(len(newX)):
    D[i,] = dist(X,newX[i])
prednum = np.argsort(D, axis = 1)[:,0:5]
```

```python
predY = np.mean(Y[prednum], axis = 1)
```

```python
plt.plot(newX,predY, c = 'red')
plt.scatter(X,Y, c = 'black')
```

```python
preds = []
k = 20
batchsize = 100
r = np.linspace(0,50,200)
for j in range(len(r)):
    newestX = r[j]
    storedX = X[:100]
    dists = dist(newestX,storedX)
    storedY = Y[:100]
    knear = np.argsort(dists)[:k]
    storeddists = dists[knear]
    storedX = X[knear]
    storedY = Y[knear]
    for i in range(int(np.ceil(len(X)/batchsize))):
        newX = X[batchsize*i:batchsize*(i+1)]
        newY = Y[batchsize*i:batchsize*(i+1)]
        newdist = dist(newX,newestX)
        newknear = np.argsort(newdist)[:k]
        newstoreddists = np.append(storeddists,newdist[newknear])
        newstoredX = np.append(storedX,newX[newknear])
        newstoredY = np.append(storedY,newY[newknear])
        index = np.argsort(newstoreddists)[:k]
        storeddists = newstoreddists[index]
        storedX = newstoredX[index]
        storedY = newstoredY[index]
    predY = np.sum(stat.norm(np.mean(storeddists),np.std(storeddists+1e-6)).pdf(storeddists)*storedY)/(np.sum(stat.norm(np.mean(storeddists),np.std(storeddists+1e-6)).pdf(storeddists))+1e-20)
    preds.append(predY)
```

```python
plt.plot(np.linspace(0,50,200),preds, c = 'red')
plt.scatter(X,Y, c = 'black')
```
