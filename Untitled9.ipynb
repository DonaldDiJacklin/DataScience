```python
X1 = np.random.randn(5000,3)
r = np.sqrt(X1[:,0]**2 + X1[:,1]**2 + X1[:,2]**2)
newX0 = X1[r < .5,]
newX1 = X1[r > 2.5,]
X = np.row_stack((newX0,newX1))
Y0 = np.zeros((len(newX0),1))
Y1 = np.ones((len(newX1),1))
Y = np.row_stack((Y0,Y1))
X = ((X - np.min(X, axis = 0))/(np.max(X,axis = 0)-np.min(X,axis= 0))) - .5
```

```python
model = NeuralNetwork(indims = 3, activations = ['tan'],nodes = [2,3], taskType = 'r')
```

```python
for i in range(1000):
    altX = X*(np.random.rand(X.shape[0], X.shape[1])<2/3)
    altX = ((altX - np.min(altX, axis = 0))/(np.max(altX,axis = 0)-np.min(altX,axis= 0)))
    model.train(altX,X, epochs = 1, lr = .01, optim = 'adam')
```

```python
newX = np.empty((len(X),2))
newX = model.layers[0].forward(X)
# for i in range(len(X)):
#     newX[i,] = model.layers[0].forward(X[i].reshape(1,-1))
```
