```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from MyML import *
```

```python
X1 = np.random.randn(1000,2)
X2 = np.random.randn(1000,2)*1.5+6
y1 = np.ones(1000)
y2 = np.zeros(1000)
full = np.row_stack((np.column_stack((X1,y1)),np.column_stack((X2,y2))))
np.random.shuffle(full)
plt.plot(X1[:,0],X1[:,1])
plt.plot(X2[:,0],X2[:,1])
```

```python
import time
```

```python
tot = TSNN(indims = 2, nodes = [10,10,1], taskType= 'c')
stocherr = []
start = time.time()
tot.train(full[:,0:2],full[:,2:], epochs = 1300, vocal = True)
end = time.time()
print(end-start)
```

```python
accuracy(tot.prediction,full[:,2:])
```

```python
model = TSNN(indims = 2, nodes = [10,10,1], taskType= 'c')
```

```python
length = 20
errlist = []
start = time.time()
for i in range(0,40):
    for j in range(0,100):
        model.train(np.matrix(full[j*length:(j+1)*length,0:2]),np.matrix(full[j*length:(j+1)*length,2:]),lr = .0001, epochs = 1,vocal = True)
#         errlist.append(model.errs[-1])
end = time.time()
print(end-start)
# plt.plot(errlist)
```

```python
accuracy(model.predict(full[:,0:2]),full[:,2:])
```
