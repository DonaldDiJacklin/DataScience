```python
import numpy as np
import MyML
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(10000,2)
```

```python
r = X[:,1]>.5
```

```python
s = X[:,0]>.5
```

```python
X1 = X[s*r]
```

```python
t = X[:,0]<.5
u = X[:,1]<.5
```

```python
X2 = X[t*u]
```

```python
Xpos = np.row_stack((X1,X2))
Xpos = np.column_stack((Xpos,np.ones(len(Xpos[:,0]))))
```

```python
X3 = X[r*t]
X4 = X[s*u]
Xneg = np.row_stack((X3,X4))
Xneg = np.column_stack((Xneg,np.zeros(len(Xneg[:,0]))))
X = np.row_stack((Xpos,Xneg))
```

```python
plt.scatter(Xpos[:,0],Xpos[:,1], s = 7)
plt.scatter(Xneg[:,0],Xneg[:,1], s = 7)
```

```python
list = []
nnmodel = MyML.NeuralNetwork(indims = 2,nodes = [10,10,1], activations = ['tan','tan'], taskType = 'c')
```

```python
Xs = X[:,0:2]
Ys = X[:,2].reshape(-1,1)
```

```python
nnmodel.layers[0].weights.shape
```

```python
nnmodel.train(Xs,Ys, epochs = 100, lr = .03,moment = .9,gamma = .8, optim = 'adam')
```

```python
MyML.accuracy(nnmodel.prediction,Ys)
```

```python
MyML.recall(nnmodel.prediction, Ys)
```

```python
MyML.confusion(nnmodel.prediction,Ys)
```

```python
np.sum(np.round(nnmodel.predict(Xs)) == Ys)
```
