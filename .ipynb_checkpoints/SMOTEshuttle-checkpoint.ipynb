```python
X1 = np.random.randn(1000,2)
X2 = np.random.randn(60,2)*[.5,.5] - .4
X = np.row_stack((X1,X2))
Y1 = np.zeros((1000,1))
Y2 = np.ones((60,1))
Y = np.row_stack((Y1,Y2))
```

```python
plt.scatter(X[:,0],X[:,1], c = Y.T.astype(int).tolist()[0])
```

```python
newX = X[np.where(Y == 1.),][0]
```

```python
D = np.empty((len(newX),len(newX)))
for i in range(len(newX)):
    D[i,] = dist(newX,newX[i])
prednum = np.argsort(D, axis = 1)[:,1:20]
```

```python
outX = newX.copy()
```

```python
for i in range(len(prednum)):
    oldpoint = newX[i]
    r = newX[prednum[i]]
    for j in range(len(r)):
        newpoint = oldpoint + np.random.uniform(low = 0,high = 1, size = 1)*(r[j,] - oldpoint)
        outX = np.row_stack((outX,newpoint))
```
