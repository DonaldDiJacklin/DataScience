```python
%%bash
sqlite3 fraud.db
create table ip_country_lookup(
lb_ip_address integer primary key, 
ub_ip_address integer not null, 
country text not null
);
create table fraud_transactions(
user_id integer primary key,
signup_time datetime not null,
purchase_time datetime not null,
purchase_value integer not null,
device_id text not null,
source text not null,
browser text not null,
sex text not null,
age integer not null,
ip_address float not null,
class integer not null
);
create table user_derived_data(
user_id integer primary key, 
ptas_seconds integer not null, 
signup_time_minutes integer, 
signup_day_of_week text, 
purchase_time_minutes integer, 
purchase_day_of_week text, 
country text
);
create table device_frequency(
device_id text primary key,
usage_count integer not null
);
create trigger populate_user_derived_data after insert
on fraud_transactions
begin
insert into user_derived_data(
user_id, 
ptas_seconds, 
signup_time_minutes, 
signup_day_of_week, 
purchase_time_minutes,
purchase_day_of_week, 
country) 
values (
new.user_id, 
(select cast ((
julianday(new.purchase_time) - julianday(new.signup_time))*24*60*60 as integer)),
(select cast ((strftime('%H', new.signup_time))*60 + strftime('%M', new.signup_time) as integer)),
(select case (cast (strftime('%w', new.signup_time) as integer))
when 0 then 'Sunday'
when 1 then 'Monday'
when 2 then 'Tuesday'
when 3 then 'Wednesday'
when 4 then 'Thursday'
when 5 then 'Friday'
else 'Saturday' end),
(select cast ((strftime('%H', new.purchase_time))*60 + strftime('%M', new.purchase_time) as integer)),
(select case (cast (strftime('%w', new.purchase_time) as integer))
when 0 then 'Sunday'
when 1 then 'Monday'
when 2 then 'Tuesday'
when 3 then 'Wednesday'
when 4 then 'Thursday'
when 5 then 'Friday'
else 'Saturday' end),
(select country from ip_country_lookup 
where new.ip_address between ip_country_lookup.lb_ip_address and ip_country_lookup.ub_ip_address)
);
insert or ignore into device_frequency(
device_id, 
usage_count)
values (
new.device_id,
0
);
update device_frequency set usage_count = usage_count + 1 where device_id is new.device_id;
end;
.mode csv
.import IpAddress_to_CountryWOH.csv ip_country_lookup
.import Fraud_DataWOH.csv fraud_transactions

select count(*) from fraud_transactions;
select count(*) from fraud_transactions where class == 0;
select count(*) from fraud_transactions where class == 1;
select count(*) from fraud_transactions natural join user_derived_data;
.headers on
.mode csv
.output combinedFraudData.csv
select * from (select * from fraud_transactions inner join device_frequency on 
device_frequency.device_id = fraud_transactions.device_id) natural join user_derived_data;
.quit
```

```python
def one_hot(df, cols):
    """
    @param df pandas DataFrame
    @param cols a list of columns to encode 
    @return a DataFrame with one-hot encoding
    """
    for each in cols:
        dummies = pd.get_dummies(df[each], prefix=each, drop_first=False)
        df = pd.concat([df, dummies], axis=1)
    return df
```

<h2><font color="red">TASK 2:</font>  Build a model to predict whether an activity is fraudulent or not. Explain how different assumptions about the cost of false positives vs. false negatives would impact the model. </h2>
<h3><font color ="blue">Generally, a false positive is called a <i>type I error</i> and a false negative is called a <i>type II error</i>.  These errors have a complementary relationship which means that there is a trade-off when optimizing one of them.  The assumptions you make about the costs associated each of these errors depends on the context of problem that the model is being applied to.  In our case, reducing false positives will increase false negatives and thus affect our fraud detection abilities.  And reducing false negatives, will increase false positives and may actually increase our losses beyond the scope of fraudulent transactions.  A false positive may inconvenience and/or alienate our "honest" customers and lose the possible revenue we would have obtained from that transaction.  A false negative is costly, because it doesn't address our original problem with fraud in the first place.  In this case, we should be using the harmonic mean of both errors to optimize our model.</font></h3>

```python
# Sigmoid function
def sigmoid(Z):
    return (1/(1+np.exp(-Z)))

# Cross Entropy cost function, scaled by the number of observations
def crossEntropy(p,y):
    return -np.sum(y*np.log(p)+(1-y)*np.log(1-p))/(len(p)+1e-10)

# Binary cross entropy cost, scaled by the number of observations
def BCEC(Y, Yhat):
    return -np.sum(np.multiply(Y,np.log(Yhat))+np.multiply((1-Y),np.log(1-Yhat)))/(len(Y[:,0]+1e-10))
```

```python
# Simple model = Logistic Regression with Elastic Net Regularization (L1 and L2)
wPred = np.random.randn(phiTrain.shape[1],1)
wIdeal = wPred
```

```python
epochs = 500
```

```python
learningRate = 1e-4
```

```python
lambda1 = .02
```

```python
lambda2 = .01
```

```python
idealValidationError = float('inf')
```

```python
trainErrors = []
validationErrors = []
epochList = []
epochCount = 0
```

```python
for i in range(epochs):
    epochCount += 1
    epochList.append(epochCount)
    pHatTrain = sigmoid(phiTrain@wPred)
    pHatValidate = sigmoid(phiValidate@wPred)
    wPred = wPred - learningRate * phiTrain.T.dot(pHatTrain-yTrain) - lambda1 * np.sign(wPred) - lambda2 * wPred
    trainError = BCEC(yTrain, pHatTrain)
    trainErrors.append(trainError)
    validationError = BCEC(yValidate, pHatValidate)
    validationErrors.append(validationError)
    if validationError < idealValidationError:
        idealValidationError = validationError
        wIdeal = wPred
plt.scatter(epochList, trainErrors, color = 'red')
plt.scatter(epochList, validationErrors, color = 'green')
red_patch = mpatches.Patch(color = 'red', label = 'Training Error')
green_patch = mpatches.Patch(color = 'green', label = 'Validation Error')
plt.legend(handles = [red_patch, green_patch], loc = 1)
plt.title("Training and Validation Errors vs. Epoch")
plt.show()
```

```python
pHatVal = sigmoid(phiValidate@wIdeal)

for i in range(yValidate.shape[1]):
    Y = yValidate
    out = pHatVal
    sortedY = Y[np.argsort(out, axis = 0)]
    
    numPos = sum(sortedY)
    numNeg = sum(1-sortedY)
    
    trueNegCount = np.cumsum(1-sortedY)
    FPR = 1 - trueNegCount/numNeg
    
    truePosCount = numPos - np.cumsum(sortedY)
    TPR = truePosCount/numPos
    
    TPR = np.insert(TPR, 0, 1, axis = 1)
    FPR = np.insert(FPR, 0, 1, axis = 1)
    
    AUC = -.5*(np.sum(TPR[:,1:]*np.diff(FPR))+np.sum(TPR[:,:-1]*np.diff(FPR)))
    accuracy = truePosCount+trueNegCount/10000
    
    plt.plot(FPR[0,:],TPR[0,:])
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title("ROC Curve")
    plt.plot([0,1],[0,1])
    auc_patch = mpatches.Patch(label = "AUC = {}".format(AUC))
    #acc_patch = mpatches.Patch(label = "Accuracy = {}".format(accuracy))
    #rec_patch = mpatches.Patch(label = "Recall = {}".format(recall))
    #plt.legend(handles=[auc_patch,acc_patch,rec_patch], loc = 1)
    plt.legend(handles=[auc_patch], loc = 4)
    plt.show()
```

```python
def makeBinaryConfusionMatrix(sortedLabels, yPred, yTrue):
    '''
    sortedLabels is a list of sorted class labels
    yPred is an np.array of binary predictions
    yTrue is an np.array of binary classes
    '''
    labels = sortedLabels.copy()
    m  = len(sortedLabels)
    columns = np.array(sortedLabelList).reshape(-1,1).T
    rows = np.array(sortedLabelList).reshape(-1,1)
    columnLabels = []
    rowLabels = []
    for label in labels:
        columnLabels.append('Actual'+str(label))
        rowLabels.append('Pred'+str(label))
    columnLabels.append('Totals')
    rowLabels.append('Totals')
    confusionMatrix = np.repeat(int(0), m * m).reshape(m,m).astype(int)
    for a in range(yPred.shape[0]):
        predValue = int(yPred[a,:][0])
        predIndex = np.where(rows==predValue)
        trueValue = int(yTrue[a,:][0])
        trueIndex = np.where(rows==trueValue)
        confusionMatrix[predIndex[0][0],trueIndex[0][0]] += 1
    
    predTotals = np.empty((confusionMatrix.shape[0],1))
    trueTotals = np.empty((1,confusionMatrix.shape[1]+1))
    for index in range(confusionMatrix.shape[0]):
        predTotals[index,:] = int(np.sum(confusionMatrix[index,:]))
        trueTotals[:,index] = int(np.sum(confusionMatrix[:,index], axis = 0))
    trueTotals[:,confusionMatrix.shape[0]] = np.sum(predTotals, axis = 0)
    confusionMatrix = np.hstack((confusionMatrix, predTotals))
    confusionMatrix = np.vstack((confusionMatrix, trueTotals))
    confusionMatrix = confusionMatrix.astype(int)
    #print(confusionMatrix)
    cMatrix = pd.DataFrame(confusionMatrix, columns=columnLabels, index=rowLabels)
    return cMatrix
```

```python
yValPred = np.rint(sigmoid(phiValidate@wIdeal))
sortedLabelList = [0,1]
confusionMatrixVal = makeBinaryConfusionMatrix(sortedLabelList, yValPred, yValidate)
print(confusionMatrixVal)
evalValMatrix = confusionMatrixVal.values
outlierValRecall = evalValMatrix[1,1]/(evalValMatrix[2,1])
#outlierValRecall = evalValMatrix[1,1]/(evalValMatrix[1,1]+evalValMatrix[0,1])
outlierValPrecision = evalValMatrix[1,1]/(evalValMatrix[1,2])
#outlierValPrecision = evalValMatrix[1,1]/(evalValMatrix[1,1]+evalValMatrix[1,0])
outlierValF1Score = (2*(outlierValRecall*outlierValPrecision))/(outlierValRecall+outlierValPrecision)
print("Model validation set outlier recall = {}".format(outlierValRecall))
print("Model validation set outlier precision = {}".format(outlierValPrecision))
print("Model validation set outlier F-1 = {}".format(outlierValF1Score))
accuracyValNum = 0
for i in range(evalValMatrix.shape[0]-1):
    accuracyValNum += evalValMatrix[i,i]
totalValAccuracy = accuracyValNum/yValidate.shape[0]
print("Total model validation accuracy = {}".format(totalValAccuracy))
```
