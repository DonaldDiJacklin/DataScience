```python
import numpy as np
import matplotlib.pyplot as plt

import time

from MyML import *
```

```python
N = 10000
D = 2
scalingFactorRed = 300
scalingFactorGreen = 55
scalingFactorBlue = 700
numPoints = N * D
addedPoints = 0
greenMatrix = np.random.normal(0, scalingFactorGreen, (N, D))
redMatrix = np.random.normal(0, scalingFactorRed, (N,D))
blueMatrix = np.random.normal(0, scalingFactorBlue, (N,D))

def makeDonut(X, lowBound, highBound):
    return np.array(list(filter(lambda x:
                                 x[0]**2+x[1]**2>lowBound**2
                                 and
                                 x[0]**2+x[1]**2<highBound**2, X)))
greenNut = makeDonut(greenMatrix, 0, 120)
redNut = makeDonut(redMatrix, 200, 275)
blueNut = makeDonut(blueMatrix, 400, 515)

plt.figure()
plt.scatter(greenNut[:,0], greenNut[:,1], c='green', s=1)
plt.scatter(redNut[:,0], redNut[:,1], c='red', s=1)
plt.scatter(blueNut[:,0], blueNut[:,1], c='blue', s=1)
```

```python
print(greenNut.shape, redNut.shape, blueNut.shape)
```

```python
greenNut = greenNut[0:839,]
redNut = redNut[0:839,]
blueNut = blueNut[0:839,]
print(greenNut.shape, redNut.shape, blueNut.shape)
```

```python
yGreen = np.hstack((np.ones((greenNut.shape[0], 1)), np.zeros((greenNut.shape[0], 1)), np.zeros((greenNut.shape[0], 1))))
yRed = np.hstack((np.zeros((redNut.shape[0], 1)), np.ones((redNut.shape[0], 1)), np.zeros((redNut.shape[0], 1))))
yBlue = np.hstack((np.zeros((blueNut.shape[0], 1)), np.zeros((blueNut.shape[0], 1)), np.ones((blueNut.shape[0], 1))))
```

```python
greenValMatrix = np.random.normal(0, scalingFactorGreen, (int(.2*N), D))
redValMatrix = np.random.normal(0, scalingFactorRed, (int(.2*N),D))
blueValMatrix = np.random.normal(0, scalingFactorBlue, (int(.2*N),D))

greenValNut = makeDonut(greenValMatrix, 0, 120)
redValNut = makeDonut(redValMatrix, 200, 275)
blueValNut = makeDonut(blueValMatrix, 400, 515)

plt.figure()
plt.scatter(greenValNut[:,0], greenValNut[:,1], c='green', s=1)
plt.scatter(redValNut[:,0], redValNut[:,1], c='red', s=1)
plt.scatter(blueValNut[:,0], blueValNut[:,1], c='blue', s=1)
```

```python
print(greenValNut.shape, redValNut.shape, blueValNut.shape)
```

```python
greenValNut = greenValNut[0:177,]
redValNut = redValNut[0:177,]
blueValNut = blueValNut[0:177,]
```

```python
yGreenVal = np.hstack((np.ones((greenValNut.shape[0], 1)), np.zeros((greenValNut.shape[0], 1)), np.zeros((greenValNut.shape[0], 1))))
yRedVal = np.hstack((np.zeros((redValNut.shape[0], 1)), np.ones((redValNut.shape[0], 1)), np.zeros((redValNut.shape[0], 1))))
yBlueVal = np.hstack((np.zeros((blueValNut.shape[0], 1)), np.zeros((blueValNut.shape[0], 1)), np.ones((blueValNut.shape[0], 1))))
```

```python
xTrain = np.vstack((greenNut, redNut, blueNut))
yTrain = np.vstack((yGreen, yRed, yBlue))

xValidate = np.vstack((greenValNut, redValNut, blueValNut))
yValidate = np.vstack((yGreenVal, yRedVal, yBlueVal))
```

```python
fullTrain = np.hstack((xTrain, yTrain))
fullValidate = np.hstack((xValidate, yValidate))
np.random.shuffle(fullTrain)
np.random.shuffle(fullValidate)
```

```python
for a in range(10):
    np.random.shuffle(fullTrain)
    np.random.shuffle(fullValidate)
```

```python
xTrain = fullTrain[:,0:2]
yTrain = fullTrain[:,2:5]

xValidate = fullValidate[:,0:2]
yValidate = fullValidate[:,2:5]
#print(xValidate.shape, yValidate.shape)
```

```python
normXTrain, xTrainColMeans, xTrainColStds = ppml.NormalizationML.zNormalize(xTrain)
normXValidate = (xValidate - xTrainColMeans)/(xTrainColStds)
```

```python
startFGDTime = time.time()
FGDmodel = ANN(xTrain.shape[1], [10,8,8,3], "tanh", True)
FGDmodel.weightInit()
FGDmodel.train(normXTrain, yTrain, 20000, .000001, normXValidate, yValidate)
endFGDTime = time.time()
FGDvalPrediction = FGDmodel.predict(normXValidate)
```

```python
totalTPFGD = 0
totalTNFGD = 0

for r in range(0, yValidate.shape[1]):
    
    truePositiveFGD = 0
    trueNegativeFGD = 0
    falsePositiveFGD = 0
    falseNegativeFGD = 0
    
    for q in range(0, yValidate.shape[0]):
        if (yValidate[q,r] == 0) & (FGDvalPrediction[q,r] == 0):
            trueNegativeFGD = trueNegativeFGD + 1
        elif (yValidate[q,r] == 0) & (FGDvalPrediction[q,r] == 1):
            falsePositiveFGD = falsePositiveFGD + 1
        elif (yValidate[q,r] == 1) & (FGDvalPrediction[q,r] == 1):
            truePositiveFGD = truePositiveFGD + 1
        elif (yValidate[q,r] == 1) & (FGDvalPrediction[q,r] == 0):
            falseNegativeFGD = falseNegativeFGD + 1
    
    accuracyFGD = (truePositiveFGD + trueNegativeFGD)/(truePositiveFGD + 
                                                       trueNegativeFGD + falsePositiveFGD + falseNegativeFGD)
    recallFGD = (truePositiveFGD)/(truePositiveFGD + falseNegativeFGD)
    
    print("Class {}:".format(r))
    print("Accuracy = {}".format(accuracyFGD))
    print("Recall = {}".format(recallFGD))
    
    totalTPFGD = totalTPFGD + truePositiveFGD
    totalTNFGD = totalTNFGD + trueNegativeFGD

print("Total validation accuracy for FGD = {}".format((totalTPFGD+totalTNFGD)/(yValidate.shape[0]*yValidate.shape[1])))
print("FGD training time = {}".format(endFGDTime - startFGDTime))
```

```python
def GCEC(Y,Yhat):
    return -np.sum(np.multiply(Y,np.log(Yhat)))/len(Y[:,0])
```

```python
SGDErrorList = []
SGDIterationList = []
iteration = 1

startSGDTime = time.time()
SGDmodel = ANN(xTrain.shape[1], [10,8,8,3], "tanh", True)
SGDmodel.weightInitialization()

batchPartition = .1
startPoint = 0
normXRowCount = normXTrain.shape[0]

for s in range(10):
    cutPoint = int(batchPartition * normXRowCount)
    partNormXTrain = normXTrain[startPoint:cutPoint,:]
    partYTrain = yTrain[startPoint:cutPoint,:]
    
    SGDIterationList.append(iteration)
    SGDmodel.train(partNormXTrain, partYTrain, partNormXTrain, partYTrain, 10000, .000001)
    SGDErrorList.append(GCEC(partYTrain, SGDmodel.probabilities))
    iteration += 1
    batchPartition += .1
    startPoint = cutPoint

endSGDTime = time.time()
```

```python
plt.scatter(SGDIterationList, SGDErrorList, color = 'green')
green_patch = mpatches.Patch(color='green', label='SGD Performance')
plt.legend(handles=[green_patch], loc=1)
plt.xlabel("Iterations")
plt.ylabel("Error")
plt.title('SGD Performance')
```
