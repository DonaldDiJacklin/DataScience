## Create some random uniform noise

```python
class1 = (np.random.rand(1000,2)-0.5)
plt.figure()
plt.scatter(class1[:,0],class1[:,1])
```

## create a donut from 2D data

```python
# Let's make a Donut:
def makeDonut(X):
    return np.array(list(filter(lambda x:
                                 x[0]**2+x[1]**2>0.2**2
                                 and
                                 x[0]**2+x[1]**2<0.4**2,class1)))
class1 = makeDonut(class1)
class1.shape
```

## Plot the donut

```python
plt.figure()
plt.scatter(class1[:,0],class1[:,1],c='r',s=3)
```

## Make a baby Gaussian in the donut

```python
class2 = np.random.randn(class1.shape[0],2)*0.04
```

## Plot both

```python
plt.figure()
plt.scatter(class2[:,0],class2[:,1],c='teal',s=2)
plt.scatter(class1[:,0],class1[:,1], c='r',s=2)
plt.xlabel('x1',fontsize = 24)
plt.ylabel('x2',fontsize = 24)
```

```python
epochs = 2000
LR = 1e-3
w = np.random.randn(phi.shape[1],1)

def sigmoid(x,w):
    return 1/(1+np.exp(-x@w))

for i in range(epochs):
    yhat = sigmoid(phi,w)
    w = w-LR*(phi.T@(yhat-y))
    #print(-np.sum(y*np.log(yhat)+(1-y)*np.log(1-yhat)))
```

To get the decision boundary equation: we do some math:
$$0=\phi w$$

In this case phi is $$\phi = [1, \sqrt{x_1^2+x_2^2}$$

So the hypothesis at the boundary becomes:
$$0 = w_0 + w_1 \sqrt{x_1^2+x_2^2}$$



```python
# Let's simulate some points to actually get this decision boundary:
N = 1e6
P = (np.random.rand(int(N),2)-0.5)
```

```python
mask = filter(lambda x: 
              int(w[0]+w[1]*np.sqrt(x[0]**2+x[1]**2)>-1e-2) 
              and  
              int(w[0]+w[1]*np.sqrt(x[0]**2+x[1]**2)<1e-2)
              ,P)

bound = np.array(list(mask))
bound.shape
```

```python
plt.figure()
plt.scatter(X[(yhat>=0.5).flatten(),0],X[(yhat>=0.5).flatten(),1],c='red',s=2)
plt.scatter(X[(yhat<=0.5).flatten(),0],X[(yhat<=0.5).flatten(),1],c='blue',s=2)
plt.scatter(bound[:,0],bound[:,1],c='k',s=1)
plt.show()
```
